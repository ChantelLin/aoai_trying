```python
import os
import pandas as pd

filename = 'twistlock_registry_3_20_24_10_38_59.csv'
df=pd.read_csv(os.path.join(os.getcwd(),filename)) # This assumes that you have placed the bill_sum_data.csv in the same directory you are running Jupyter Notebooks
df
```

```python
df_bills = df[['Fix Status', 'Risk Factors', 'Description']]
df_bills
```

```python
import re

pd.options.mode.chained_assignment = None #https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#evaluation-order-matters

# s is input text
def normalize_text(s, sep_token = " \n "):
    s = re.sub(r'\s+',  ' ', s).strip()
    s = re.sub(r". ,","",s)
    # remove all instances of multiple spaces
    s = s.replace("..",".")
    s = s.replace(". .",".")
    s = s.replace("\n", "")
    s = s.strip()
    
    return s

df_bills['Description']= df_bills["Description"].apply(lambda x : normalize_text(x))
```

```python
import tiktoken

tokenizer = tiktoken.get_encoding("cl100k_base")
df_bills['n_tokens'] = df_bills["Description"].apply(lambda x: len(tokenizer.encode(x)))
df_bills = df_bills[df_bills.n_tokens<8192]
len(df_bills)
```

```python
from openai import AzureOpenAI

client = AzureOpenAI(
    api_key = api_key,
    api_version = api_version,
    azure_endpoint = api_base
    )

def generate_embeddings(text, model=embedding_deploy_name): # model = "deployment_name"
    return client.embeddings.create(input = [text], model=model).data[0].embedding
```

```python
df_bills['ada_v2'] = df_bills["Description"].apply(lambda x : generate_embeddings (x)) # model should be set to the deployment name you chose when you deployed the text-embedding-ada-002 (Version 2) model
```

```python
import numpy as np

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def get_embedding(text, model=embedding_deploy_name): # model = "deployment_name"
    return client.embeddings.create(input = [text], model=model).data[0].embedding

def search_docs(df, user_query, top_n=4, to_print=True):
    embedding = get_embedding(
        user_query,
        model = embedding_deploy_name # model should be set to the deployment name you chose when you deployed the text-embedding-ada-002 (Version 2) model
    )
    df["similarities"] = df.ada_v2.apply(lambda x: cosine_similarity(x, embedding))

    res = (
        df.sort_values("similarities", ascending=False)
        .head(top_n)
    )
    if to_print:
        display(res)
    return res

```

```python
res = search_docs(
    df_bills, 
    "Can I get information which is high in risk factor?", 
    top_n=4)
```
